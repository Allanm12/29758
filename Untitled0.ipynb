```python
# load iris data from sklearn
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
imported = load_iris()

# create a pandas Dataframe from the data
iris_df = pd.DataFrame(imported.data, columns=imported.feature_names)

# Add the target variable to the dataframe
iris_df['species'] = imported.target

#Dispaly the firdt few rows of the dataframe
print(iris_df.head())
```

```python
iris_df.sample(10)
```

```python
# Features (x) and target (Y)
x = iris_df[["petal length (cm)","sepal width (cm)","petal width (cm)"]]
y = iris_df["sepal length (cm)"]
```

```python
# import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# y =mx=
```

```python
# model selection
mymodel = LinearRegression()

# model training
mymodel.fit(x_train, y_train)

# model evaluation
Y_pred = mymodel.predict(x_test)
```

```python
# evaluate the model
mse = mean_squared_error(y_test, Y_pred)
rmse = mean_squared_error(y_test, Y_pred, squared=False)
r2 = r2_score(y_test, Y_pred)

print(f"Mean Squared Error(MSE): {mse}")
print(f"Root Mean Squared Error(rmse):, {rmse}")
print(f"R-squared (R2): {r2}")
```

```python
def core(dataset, threshold):
  col_corr = set()
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i, j]) > threshold:
        colname = corr_matrix.columns[i]
        col_corr.add(colname)
  return col_corr
```
